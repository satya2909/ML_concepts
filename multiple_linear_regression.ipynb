```python
import pandas as pd
import numpy as np
```

```python
stu = pd.read_csv('/content/drive/MyDrive/datasets/Student_Performance.csv')
stu.head()
```

```python
x = stu[['Hours Studied','Previous Scores','Sleep Hours','Sample Question Papers Practiced']].copy()
y = stu[['Performance Index']].copy()
y
```

```python
x['intercepts']=1
x = x[['intercepts','Hours Studied','Previous Scores','Sleep Hours','Sample Question Papers Practiced']]
x
```

```python
x_T= x.T
x_T
```

```python
B = np.linalg.inv(x_T @ x) @ x_T @ y
B.index = x.columns
B
```

```python
new_performances = x @ B
new_performances
```

```python
ssr = ((y-new_performances)**2).sum() # it is calculating the error
sst = ((y-y.mean())**2).sum() # it is the variance
r_squared = 1 - ssr/sst
print(r_squared)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

lr = LinearRegression()
lr.fit(stu[['Hours Studied','Previous Scores','Sleep Hours','Sample Question Papers Practiced']],stu[['Performance Index']])

lr.coef_
```
